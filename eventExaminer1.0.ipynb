#### Functions:

```python
#Function that recognizes 'Yes' and 'Y' strings and returns true if string is recognized, these aren't case sensitive   
def isYes(s):
    s = s.lower() 
    if s == 'yes' or s == 'y' or s == 'true':  
        return True
    else:
        return False
    
#Function that converts a date string into a datetime object (Strings must be in the following format YYYY-MM-DD)
def stringToDate(x):    
    x=str(x)
    y=int(x[0:4])
    m=int(x[5:7])
    d=int(x[8:10])   
    return datetime.datetime(y,m,d)
#eg 
#2012-12-01 00:00:00
#2013-01-15 00:00:00

#Thousand separator function
def thouSeparator(x, pos):  # formatter function takes tick label and tick position       
    max_y = max(plt.ylim())
    min_y = min(plt.ylim())
    if max_y >= 1000 or min_y <= -1000:           
        s = str('{:,g}'.format(x))                                
        return s 
    else: 
        s = '{:g}'.format(x)
        return s
    
#Function that removes the numbers from node names. Used for title formatting
def stripNumbersFromNodes(x):
    x = x[:3]
    return x

#Function that removes last character from a string. Used for title formatting
def stripLastChar(x):
    x = x[:-1]
    return x

#Function that extracts the day from a date string (removes the 0). Used for title formatting
def getDay(x):
    x = str(x)    
    d = int(x[8:10])
    if int(x[8])==0:
        d = int(x[9:10])
    return str(d)  

#Function that extracts the month number from date string (will remove the 0). Use dateBeg or dateEnd as arguments. Used for title formatting
def getMonthNumber(x):
    x = str(x)    
    m = int(x[5:7])
    if int(x[5])==0:
        m = int(x[6:7])
    return str(m)  

#Function that extracts the year from a date string. Use dateBeg or dateEnd as arguments. Used for title formatting
def getYear(x):
    #x = str(x)
    year = int(x[0:4])
    return str(year) 

#Function that returns the full name of the month as a string. Use dateBeg or dateEnd as arguments. Used for title formatting
def getMonth(x):
    #x = str(x)
    m = int(x[5:7])
    month = ''
    if m == 1:
        month = 'January'
    elif m == 2:
        month = 'February'
    elif m == 3:
        month = 'March'
    elif m == 4:
        month = 'April'
    elif m == 5:
        month = 'May'
    elif m == 6:
        month = 'June'
    elif m == 7:
        month = 'July'
    elif m == 8:
        month = 'August'
    elif m == 9:
        month = 'September'
    elif m == 10:
        month = 'October'
    elif m == 11:
        month = 'November'
    elif m == 12:
        month = 'December'    
    return month    

#Function that returns the shorthand name of the month as a string. Use dateBeg or dateEnd as arguments. Used for title formatting
def getMonthShortName(x):
    #x = str(x)
    m = int(x[5:7])
    month = ''
    if m == 1:
        month = 'Jan'
    elif m == 2:
        month = 'Feb'
    elif m == 3:
        month = 'Mar'
    elif m == 4:
        month = 'Apr'
    elif m == 5:
        month = 'May'
    elif m == 6:
        month = 'Jun'
    elif m == 7:
        month = 'Jul'
    elif m == 8:
        month = 'Aug'
    elif m == 9:
        month = 'Sep'
    elif m == 10:
        month = 'Oct'
    elif m == 11:
        month = 'Nov'
    elif m == 12:
        month = 'Dec'    
    return month  
    
def getHours(x):      
    h = x[0:2] 
    return h
#eg: print getHours(tradePeriods[0]) = 00

def getMinutes(x):
    m = x[3:5]
    return m
#eg: print getMinutes(tradePeriods[0]) = 00

#Function for building graph titles for the PDF report
def getDateForTitle(dateB,dateE):    
    dateForTitle = ''      
    if getDay(dateB) == getDay(dateE) and getMonth(dateB) == getMonth(dateE) and getYear(dateB) == getYear(dateE):
        dateForTitle = getDay(dateB) + ' ' + getMonth(dateB) + ' ' + getYear(dateB)
    elif getMonth(dateB) == getMonth(dateE) and getYear(dateB) == getYear(dateE):
        dateForTitle = getDay(dateB) + '-' + getDay(dateE) + ' ' + getMonth(dateB) + ' ' + getYear(dateB)
    elif getYear(dateB) == getYear(dateE):
        #dateForTitle = getDay(dateB) + ' ' + getMonth(dateB) + ' - ' + getDay(dateE) + ' ' + getMonth(dateE) + ' ' + getYear(dateB)
        dateForTitle = getDay(dateB) + ' ' + getMonth(dateB) + ' '+u'\u2014'+' ' + getDay(dateE) + ' ' + getMonth(dateE) + ' ' + getYear(dateB)
    else:
        #dateForTitle = getDay(dateB) + ' ' + getMonth(dateB) + ' ' + getYear(dateB) + ' - ' + getDay(dateE) + ' ' + getMonth(dateE) + ' ' + getYear(dateE)
        dateForTitle = getDay(dateB) + ' ' + getMonth(dateB) + ' ' + getYear(dateB) + ' '+u'\u2014'+' ' + getDay(dateE) + ' ' + getMonth(dateE) + ' ' + getYear(dateE)    
    return dateForTitle

print 'Day begin:',getDay(dateBeg)
print 'Month begin:',getMonth(dateBeg)
print 'Day end:',getDay(dateEnd)
print 'Month end:',getMonth(dateEnd)
print 'Year begin:',getYear(dateBeg)
print 'Year end:',getYear(dateEnd)
print 'Date used for title:',getDateForTitle(dateBeg,dateEnd)
```

```python
#PAGE_HEIGHT=defaultPageSize[1]; PAGE_WIDTH=defaultPageSize[0]
#styles = getSampleStyleSheet()

class PdfImage(Flowable):
    def __init__(self, filename_or_object, width=None, height=None, kind='direct'):
        from reportlab.lib.units import inch
        # If using StringIO buffer, set pointer to begining
        if hasattr(filename_or_object, 'read'):
            filename_or_object.seek(0)
        page = PdfReader(filename_or_object, decompress=False).pages[0]
        self.xobj = pagexobj(page)
        self.imageWidth = width
        self.imageHeight = height
        x1, y1, x2, y2 = self.xobj.BBox

        self._w, self._h = x2 - x1, y2 - y1
        if not self.imageWidth:
            self.imageWidth = self._w
        if not self.imageHeight:
            self.imageHeight = self._h
        self.__ratio = float(self.imageWidth)/self.imageHeight
        if kind in ['direct','absolute'] or width==None or height==None:
            self.drawWidth = width or self.imageWidth
            self.drawHeight = height or self.imageHeight
        elif kind in ['bound','proportional']:
            factor = min(float(width)/self._w,float(height)/self._h)
            self.drawWidth = self._w*factor
            self.drawHeight = self._h*factor

    def wrap(self, aW, aH):
        return self.drawWidth, self.drawHeight

    def drawOn(self, canv, x, y, _sW=0):
        if _sW > 0 and hasattr(self, 'hAlign'):
            a = self.hAlign
            if a in ('CENTER', 'CENTRE', TA_CENTER):
                x += 0.5*_sW
            elif a in ('RIGHT', TA_RIGHT):
                x += _sW
            elif a not in ('LEFT', TA_LEFT):
                raise ValueError("Bad hAlign value " + str(a))

        xobj = self.xobj
        xobj_name = makerl(canv._doc, xobj)

        xscale = self.drawWidth/self._w
        yscale = self.drawHeight/self._h

        x -= xobj.BBox[0] * xscale
        y -= xobj.BBox[1] * yscale

        canv.saveState()
        canv.translate(x, y)
        canv.scale(xscale, yscale)
        canv.doForm(xobj_name)
        canv.restoreState()
```
